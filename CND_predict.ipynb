#Import data
You need to import both the image and the model


```
from keras.models import load_model, Sequential
import cv2
import matplotlib.pyplot as plt
from google.colab import files
from keras.preprocessing.image import load_img
import numpy as np

files.upload() #upload your picture
files.upload() #upload the model
```

#Prediction

```
FILE_PATH=""  #enter picture path

CATEGORIES=["cat","dog"]
MODEL_PATH=""  #enter model path
IMAGE_SIZE=(128,128)
def prepare(filepath):
    img_array=cv2.imread(filepath)
    new_array= cv2.resize(img_array, IMAGE_SIZE)
    return new_array.reshape(-1, IMAGE_SIZE[0], IMAGE_SIZE[1],3)

model = load_model(MODEL_PATH)

prediction = model.predict([prepare(FILE_PATH)])
  
img = load_img(FILE_PATH, target_size=IMAGE_SIZE)
plt.imshow(img)
plt.xlabel("This is a {}".format(CATEGORIES[int(np.argmax(prediction))]))
plt.show()
```
